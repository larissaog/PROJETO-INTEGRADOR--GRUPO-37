```python
import pandas as pd

reviews = pd.read_csv("p5_order_reviews_dataset.csv")
orders = pd.read_csv("p5_orders_dataset.csv")
items = pd.read_csv("p5_order_items_dataset.csv")
products = pd.read_csv("p5_products_dataset.csv")
```

```python
df_full = reviews.merge(orders, on="order_id", how="left")
df_full = df_full.merge(items, on="order_id", how="left")
df_full = df_full.merge(products, on="product_id", how="left")
```

```python
# Remover linhas sem nota
df_full = df_full.dropna(subset=["review_score"])

# Converter datas
df_full["order_delivered_customer_date"] = pd.to_datetime(
    df_full["order_delivered_customer_date"], errors="coerce"
)

df_full["order_estimated_delivery_date"] = pd.to_datetime(
    df_full["order_estimated_delivery_date"], errors="coerce"
)

# Criar delivery_delay
df_full["delivery_delay"] = (
    df_full["order_delivered_customer_date"] - 
    df_full["order_estimated_delivery_date"]
).dt.days

# Remover linhas sem delay
df_full = df_full.dropna(subset=["delivery_delay"])

# Filtrar valores extremos
df_full = df_full[df_full["delivery_delay"].between(-30, 60)]

# Salvar dataset limpo
df_full.to_csv("dados_limpos.csv", index=False)
```
