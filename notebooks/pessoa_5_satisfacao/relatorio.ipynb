# 7. Insights e recomendações
# -----------------------------

INSIGHTS:
1. A maioria das avaliações é 5, mas há um percentual relevante de notas 1-3.
2. Existe correlação entre atrasos na entrega e notas baixas.
3. Certas categorias de produtos apresentam mais avaliações negativas (ex: 'beleza', 'cama mesa e banho').

RECOMENDAÇÕES:
- Melhorar logística e cumprimento de prazos de entrega.
- Revisar categorias com maior índice de insatisfação e analisar problemas específicos.
- Monitorar pedidos atrasados e notificar clientes proativamente.
