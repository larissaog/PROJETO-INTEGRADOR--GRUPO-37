```python
import pandas as pd
import plotly.express as px

# O segredo: carregar o arquivo que já saiu LIMPO do notebook 02
caminho = "../../data/cleaned/pessoa_2/produtos.csv"
df = pd.read_csv(caminho)

print(f"✅ Sucesso! {len(df)} registros carregados para análise visual.")
df.head() # Só para conferir se as colunas estão em português
```

```python
# Matriz de Correlação e Insights 

# No Notebook 03, usamos o nome 'df'
colunas_analise = ['Faturamento', 'Frete', 'Peso', 'Avaliacao']

# Criando a matriz (calcula a relação entre os números)
matriz_correlacao = df[colunas_analise].corr()

# Exibindo o relatório
print("RELATÓRIO DE CORRELAÇÕES RELEVANTES:")
print("-" * 50)
display(matriz_correlacao)
print("-" * 50)

# Verificação automática do Peso vs Frete
corr_peso_frete = matriz_correlacao.loc['Peso', 'Frete']

print(f"Insight: A correlação entre Peso e Frete é de {corr_peso_frete:.2f}")

if corr_peso_frete > 0.7:
    print("Conclusão: Existe uma forte relação. O peso é o principal fator de custo do frete.")
elif corr_peso_frete > 0.4:
    print("Conclusão: Existe uma relação moderada entre peso e custo de frete.")
else:
    print("Conclusão: O peso não é o único fator determinante para o valor do frete.")
```

```python
# Gráfico 1: Top 10 Faturamento
top_fat = df.groupby('Categoria')['Faturamento'].sum().sort_values(ascending=False).head(10).reset_index()
fig1 = px.bar(top_fat, x='Categoria', y='Faturamento', title="Top 10 Categorias por Faturamento", text_auto='.2s')
fig1.show()

# Gráfico 2: Top 10 Volume (Vendas)
top_vol = df['Categoria'].value_counts().head(10).reset_index()
top_vol.columns = ['Categoria', 'Quantidade']
fig2 = px.bar(top_vol, x='Quantidade', y='Categoria', orientation='h', title="Top 10 Categorias por Volume", color='Quantidade')
fig2.show()
```

```python
# Gráfico 5: Pizza (Mix de Produtos)
fig5 = px.pie(top_vol, values='Quantidade', names='Categoria', title="Mix de Produtos (Distribuição)", hole=0.4)
fig5.show()
```
