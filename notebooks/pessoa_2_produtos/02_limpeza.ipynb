```python
import warnings
warnings.filterwarnings('ignore')

#Nesta etapa são importadas e carregadas as bibliotecas necessárias para manipulação e análise dos dados.

import pandas as pd
import numpy as np

df_itens = pd.read_csv("../../data/cleaned/pessoa_2/"'itens.csv')
df_produtos = pd.read_csv("../../data/cleaned/pessoa_2/"'produtos.csv')
df_traducao = pd.read_csv('../../data/cleaned/pessoa_2/traducao.csv')
df_avaliacoes = pd.read_csv("../../data/cleaned/pessoa_2/"'avaliacoes.csv')
```

```python
# Tratamento de dados 

# 1. Verificamos as colunas que realmente existem
colunas_prod = df_produtos.columns.tolist()
colunas_itens = df_itens.columns.tolist()

# 2. Só fazemos o merge se o 'product_id' existir nos dois
if 'product_id' in colunas_prod and 'product_id' in colunas_itens:
    vendas_base = pd.merge(df_itens, df_produtos, on='product_id', how='left')
else:
    # Se não existir, significa que os dados já podem estar unidos ou o nome mudou
    print("Aviso: 'product_id' não encontrado. Usando base de produtos direta.")
    vendas_base = df_produtos

# 3. Tentamos unir com a tradução (se a coluna de categoria existir)
if 'product_category_name' in vendas_base.columns:
    vendas_traduzidas = pd.merge(vendas_base, df_traducao, on='product_category_name', how='left')
else:
    vendas_traduzidas = vendas_base

# 4. Filtro Final (Usando apenas o que o arquivo tiver)
mapa_colunas = {
    'product_category_name_english': 'Categoria',
    'Categoria': 'Categoria', # Caso já esteja traduzido
    'price': 'Faturamento',
    'Faturamento': 'Faturamento',
    'freight_value': 'Frete',
    'Frete': 'Frete',
    'review_score': 'Avaliacao',
    'Avaliacao': 'Avaliacao'
}

# Filtra apenas as colunas que estão na tabela
existentes = [col for col in mapa_colunas.keys() if col in vendas_traduzidas.columns]
df_analise = vendas_traduzidas[existentes].copy()

# Renomeia para o padrão final
df_analise.rename(columns=mapa_colunas, inplace=True)

# 5. Salva o arquivo final
caminho_final = "../../data/cleaned/pessoa_2/produtos_limpos.csv"
df_analise.to_csv(caminho_final, index=False)

print("✅ Limpeza concluída! Arquivo salvo para o Notebook 03.")
```
